```julia
using Plots
using LinearAlgebra

# Define the original ISAC signal (as a 2D vector for simplicity)
signal = [1.0, 1.0]  # Original signal in a 2D space

# Define the deterministic subspace (basis vector)
deterministic_basis = [1.0, 0.0]  # Basis for deterministic subspace (aligned with x-axis)

# Define the random subspace (orthogonal to deterministic)
random_basis = [0.0, 1.0]  # Basis for random subspace (aligned with y-axis)

# Create projection matrices
P_d = deterministic_basis * deterministic_basis' / dot(deterministic_basis, deterministic_basis)  # Deterministic projector
P_r = random_basis * random_basis' / dot(random_basis, random_basis)  # Random projector

# Project the signal onto each subspace
deterministic_component = P_d * signal
random_component = P_r * signal

# Verify decomposition (signal = sum of components)
reconstructed_signal = deterministic_component + random_component

# Plot the original signal and its components
plot(
    [-0.5, 1.5], [-0.5, 1.5]
    ; seriestype=:line
    , label=:false, legend=:bottomright
    , xlabel="Dimension 1", ylabel="Dimension 2"
    , title="Projector Metaphor for ISAC Signal"
    , xlims = (-1,3)
)

# Plot deterministic basis (x-axis)
plot!([0, deterministic_basis[1]], [0, deterministic_basis[2]], color=:blue, label="Deterministic Subspace", arrow=:true)

# Plot random basis (y-axis)
plot!([0, random_basis[1]], [0, random_basis[2]], color=:green, label="Random Subspace", arrow=:true)

# Plot original signal
plot!([0, signal[1]], [0, signal[2]], color=:black, linestyle=:solid, label="Original Signal", arrow=:true)

# Plot deterministic component
plot!([0, deterministic_component[1]], [0, deterministic_component[2]], color=:blue, linestyle=:dash, label="Deterministic Component", arrow=:true)

# Plot random component
plot!([0, random_component[1]], [0, random_component[2]], color=:green, linestyle=:dash, label="Random Component", arrow=:true)

# Plot reconstructed signal for confirmation (it will overlap with the original)
plot!([0, reconstructed_signal[1]], [0, reconstructed_signal[2]], color=:red, linestyle=:dashdot, label="Reconstructed Signal")
```
