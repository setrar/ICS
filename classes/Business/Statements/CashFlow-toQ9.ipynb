```julia
using Plots
```

```julia

# Define the data
quarters = ["Q1", "Q2", "Q3", "Q4"] # , "Q5", "Q6", "Q7", "Q8", "Q9"]
ss_revenue = [0, 1.26e6, 3.78e6, 11.34e6]
ds_revenue = [0, 2.1e6, 6.3e6, 12.6e6]
total_revenue = [0, 3.36e6, 10.08e6, 23.94e6]
cogs_ss = [0, 540800, 1081600, 2163200]
cogs_ds = [0, 660000, 1320000, 2640000]
total_cogs = [0, 1200800, 2401600, 4803200]
gross_profit = [860000, 2159200, 8879200, 22739200]
opex = [20000, 500000, 1000000, 2000000]
net_income_loss = [840000, 1659200, 7879200, 20739200]
```

```julia


# Plotting
p = plot(quarters, [ss_revenue, ds_revenue, total_revenue, cogs_ss, cogs_ds, total_cogs, gross_profit, opex, net_income_loss],
    label=["Standard Sales" "Deluxe Sales" "Total Revenue" "COGS (Standard Sales)" "COGS (Deluxe Sales)" "Total COGS" "Gross Profit" "OpEx" "Net Income (Loss)"],
    colour=[:blue :orange :green :red :purple :brown :pink :gray :black],
    marker=([:circle :square :utriangle :diamond :cross :star :hexagon :octagon :pentagon]),
    linestyle=([:solid :solid :solid :dash :dash :dash :dot :dashdot :dashdotdot]),
    title="Financial Overview by Quarter (CHF)",
    xlabel="Quarter", ylabel="Amount (CHF)",
    legend=:topleft)

# savefig(p, "images/financial_projection_Q9.png");
# Display the plot
display(p)
```
