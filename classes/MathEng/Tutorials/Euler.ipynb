````julia
using Plots
````

````julia
# Parameters
ω = 2π          # Angular frequency (1 Hz)
t = 0:0.01:2    # Time vector from 0 to 2 seconds
j = √(Complex(-1.0))
````

````julia
# Euler's formula: e^(jωt) = cos(ωt) + j*sin(ωt)
# signal = exp.(j * ω .* t)
signal = cos.(ω .* t) + j*sin.(ω .* t);
````

````julia
# Plot real and imaginary parts
plot(t, real.(signal)
    , xlabel = "Time (s)", ylabel = "Amplitude"
    , title = "Signal using Euler's Formula"
    , label="Re{e^{jωt}}", lw=2
)
plot!(t, imag.(signal), label="Im{e^{jωt}}", lw=2)
````

````julia
# Plot in complex plane (top-down view)
plot(real.(signal), imag.(signal)
     , label="e^{jωt} Trajectory"
     , xlabel="Real", ylabel="Imaginary"
     , title="Top-Down View of e^{jωt} on the Complex Plane"
     , aspect_ratio=:equal, legend=:bottomright
     , lw=2)
````

````julia
# Extract components
x = t; y = real.(signal); z = imag.(signal);
````

````julia
# 3D Plot of the signal
plt = plot(x, y, z
    , label="e^{jωt}"
    , xlabel="Time (s)", ylabel="Real Part", zlabel="Imaginary Part"
    , title="3D View of e^{jωt} with Zero Axis"
    , legend=:topright, lw=4
    , camera=(45, 30))

# Add line traversing through origin along time
plot!(plt, [t[1], t[end]], [0, 0], [0, 0],
    label="Zero Line",
    lw=2,
    linestyle=:dash,
    color=:black)
````

````julia
# 3D Plot of the signal
plt = plot(x, y, z
    , label="e^{jωt}"
    , xlabel="Time (s)", ylabel="Real Part", zlabel="Imaginary Part"
    , title="3D View of e^{jωt} with Zero Axis"
    , legend=:topright, lw=4
    , camera=(60, 30))

# Add line traversing through origin along time
plot!(plt, [t[1], t[end]], [0, 0], [0, 0],
    label="Zero Line",
    lw=2,
    linestyle=:dash,
    color=:black)
````

**Sinusoidal signals look like a coil in 3D** — **only specific combinations do**, especially **complex exponentials** or **paired sinusoids with quadrature phase**.

---

### 🧠 Let’s break it down:

#### ✅ When you **do** get a coil in 3D:

You get a **3D helix (coil)** when you plot a signal like:

```julia
x(t) = t
y(t) = cos(ωt)
z(t) = sin(ωt)
```

This forms a **helix** because the signal traces a **circular path in the y-z plane** (via `cos` and `sin`), and moves **linearly in x (time)**.

This pattern arises in:

* **Euler's formula**: $e^{j\omega t} = \cos(\omega t) + j \sin(\omega t)$
* **Quadrature carriers** in communications (I/Q modulation)
* **Spinning phasors** in signal processing
* Circular/elliptical motion physics

---

### ❌ When you **don’t** get a coil:

#### Case 1: **Pure real sinusoid**

```julia
x(t) = t
y(t) = sin(ωt)
z(t) = 0
```

You just get a **wave along x-y**, flat in z → no coil.

#### Case 2: **Arbitrary sinusoids not in quadrature**

```julia
y(t) = sin(ωt)
z(t) = sin(ωt + φ)   # not exactly ±π/2
```

You get **wobbly curves**, possibly **Lissajous figures** — not a helix.

#### Case 3: **Mixed frequencies**

```julia
y(t) = sin(ω₁t)
z(t) = sin(ω₂t)
```

The signal may trace **complex 3D shapes**, but not a simple coil.

---

### 🎯 Summary:

| Signal Type                        | 3D Shape            |
| ---------------------------------- | ------------------- |
| `cos(ωt) + j*sin(ωt)`              | Helix / Coil        |
| Pure `sin(ωt)` or `cos(ωt)`        | Flat wave           |
| `sin(ωt)`, `sin(2ωt)`              | Complex waveform    |
| `sin(ωt)`, `cos(ωt + φ)` (φ ≠ π/2) | Lissajous-like path |
