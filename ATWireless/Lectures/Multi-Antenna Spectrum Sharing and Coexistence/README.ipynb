### CBRS 

The term you're referring to is **Citizens Broadband Radio Service (CBRS)**. CBRS is a 150 MHz wide broadcast band in the 3.5 GHz spectrum (3550 MHz to 3700 MHz) in the United States. This band was previously reserved for military use but has been repurposed to allow shared access among federal and non-federal users, including private companies. 

The Federal Communications Commission (FCC) established rules for commercial use of this band in 2017, enabling wireless carriers to deploy 5G networks without acquiring traditional spectrum licenses. This initiative aims to promote more efficient use of the spectrum and foster innovation in wireless services. 

CBRS operates under a three-tiered spectrum sharing framework:

1. **Incumbent Access**: Reserved for existing federal users, primarily the U.S. Navy, which retains priority access.
2. **Priority Access License (PAL)**: Licenses auctioned to commercial users for exclusive use in specific geographic areas.
3. **General Authorized Access (GAA)**: Open to the public for unlicensed use, provided it doesn't interfere with higher-tier users.

This structure ensures that the spectrum is utilized efficiently while protecting incumbent military operations. 

For more detailed information, you can refer to the FCC's overview of the 3.5 GHz Band and the CBRS Alliance's resources. 

## New Radio-Unlicensed

**NR-U Coexistence** refers to the mechanisms and strategies used to enable **5G New Radio-Unlicensed (NR-U)** technology to coexist with other technologies, such as Wi-Fi, that also operate in unlicensed spectrum bands.

NR-U is part of the 3GPP standardization efforts for 5G, designed to extend the capabilities of 5G into unlicensed frequency bands like **5 GHz** and **6 GHz**. Since these bands are shared with other technologies, coexistence mechanisms are critical to prevent interference and ensure fair and efficient use of the spectrum.

---

### **Key Concepts of NR-U Coexistence**

1. **Unlicensed Spectrum Usage**:
   - NR-U allows 5G to operate in bands not exclusively licensed for mobile network operators.
   - Examples include the **5 GHz** (U-NII bands) and **6 GHz** (Wi-Fi 6E) bands.

2. **Shared Spectrum Access**:
   - Devices and networks operating in unlicensed bands must use coexistence mechanisms to avoid disrupting each other.
   - Common coexisting technologies: **Wi-Fi, NR-U, LTE-U, and LAA (Licensed Assisted Access).**

3. **Regulatory Requirements**:
   - Unlicensed spectrum use is subject to regulatory constraints that mandate coexistence.
   - For example, regions like the EU require **Listen Before Talk (LBT)**, which ensures a device senses the channel before transmitting.

4. **Fairness and Efficiency**:
   - NR-U incorporates mechanisms to maintain fairness among technologies like Wi-Fi and NR-U to share spectrum without one dominating the other.
   - Techniques like time-sharing and dynamic frequency selection are applied.

---

### **NR-U Coexistence Mechanisms**

1. **Listen Before Talk (LBT)**:
   - Before transmitting, devices check the spectrum to ensure it is free.
   - If the channel is busy, the device waits for a random backoff time before retrying.

2. **Channel Access Priority Classes**:
   - Different traffic types are assigned different priorities.
   - High-priority traffic (e.g., real-time voice/video) gets quicker access to the channel.

3. **Adaptive Duty Cycling**:
   - Adjusts transmission and idle times to balance usage among technologies.

4. **Dynamic Frequency Selection (DFS)**:
   - Ensures devices avoid radar and other high-priority services that may operate in the same band.

5. **Power Control**:
   - Limits transmission power to reduce interference in shared spectrum environments.

6. **Spectrum Aggregation**:
   - Combines licensed and unlicensed spectrum to improve capacity while ensuring compatibility.

---

### **Applications of NR-U Coexistence**
- **Enhanced Mobile Broadband (eMBB)**: Extending 5G coverage and capacity in dense areas.
- **IoT Devices**: Supporting low-latency, high-density IoT deployments.
- **Enterprise Networks**: NR-U enables private 5G networks using unlicensed bands for industrial automation or campus-wide connectivity.

---

### **Challenges**
1. **Spectrum Crowding**: High usage of unlicensed bands, especially with Wi-Fi 6 and 6E, can lead to contention.
2. **Fair Coexistence**: Ensuring fair access for Wi-Fi and NR-U without degrading performance for either.
3. **Interference Mitigation**: Managing cross-technology interference effectively.

The concept of **Single Input Multiple Output (SIMO)** in wireless communications, where a single antenna transmits to multiple receiving antennas, has evolved through contributions from various researchers over time. While it's challenging to attribute its invention to a single individual, several key figures have significantly advanced this field:

1. **Claude Shannon**: In 1948, Shannon introduced the foundational principles of information theory, establishing the theoretical limits of communication systems, which underpin the development of technologies like SIMO.

2. **Jack K. Wolf and David Slepian**: In the 1970s, they conducted research on multiple-antenna systems, exploring the capacity and performance benefits of such configurations.

3. **Arogyaswami Paulraj**: A professor at Stanford University, Paulraj is renowned for his pioneering work in multiple-antenna systems, including MIMO (Multiple Input Multiple Output) technologies. His research has been instrumental in advancing wireless communication systems.

While SIMO systems are a subset of MIMO technologies, the development of SIMO has been a collaborative effort within the scientific community, building upon the foundational work of these and other researchers. 

### **Comparison: TDD vs. FDD**

| **Feature**             | **TDD (Time Division Duplexing)**                | **FDD (Frequency Division Duplexing)**          |
|--------------------------|--------------------------------------------------|-------------------------------------------------|
| **Frequency Usage**      | Single frequency for both uplink (UL) and downlink (DL), separated by time. | Separate frequencies for UL and DL.             |
| **Traffic Adaptation**   | Flexible allocation of UL/DL time slots based on demand. | Fixed UL/DL bandwidth, cannot adapt dynamically. |
| **Hardware Complexity**  | Lower: Requires only one frequency band.         | Higher: Requires separate frequency bands and duplexers. |
| **Latency**              | Slightly higher due to time-slot switching.      | Lower as UL and DL occur simultaneously.        |
| **Spectrum Efficiency**  | More efficient in asymmetric traffic scenarios.  | More efficient for symmetric traffic.           |
| **Synchronization**      | Requires strict timing coordination between devices. | No timing synchronization required.             |
| **Interference**         | Prone to interference between adjacent cells if unsynchronized. | Less interference between cells due to separate UL/DL bands. |
| **Ideal Use Case**       | Asymmetric traffic: data-heavy applications (e.g., 5G, IoT, video streaming). | Symmetric traffic: traditional voice calls or data services. |
| **Deployment Cost**      | Lower due to single-frequency usage and simpler hardware. | Higher due to the need for separate frequency bands and duplexing filters. |
| **Bandwidth Requirement**| Can operate in narrower bandwidths.             | Requires larger spectrum blocks for separate UL/DL. |
| **Examples**             | Wi-Fi, LTE-TDD, 5G NR-TDD, WiMAX.               | GSM, LTE-FDD, WCDMA, traditional telephony systems. |

---

### **Key Differences**
1. **Flexibility**: 
   - **TDD** allows dynamic UL/DL allocation, making it better suited for data-heavy applications where downlink often dominates.
   - **FDD** is rigid, with fixed uplink and downlink allocations.

2. **Spectrum Use**:
   - **TDD** uses spectrum more efficiently by sharing a single frequency for UL and DL.
   - **FDD** requires paired frequencies, which can be challenging in spectrum-constrained environments.

3. **Latency**:
   - **TDD** introduces switching delays, while **FDD** enables simultaneous communication, reducing latency.

4. **Cost**:
   - **TDD** is cheaper to deploy and maintain, especially for systems with constrained resources.

---

### **When to Use Which**
- **TDD**: 
   - Ideal for modern applications like 5G, IoT, and high-data environments where asymmetric traffic patterns are common.
   - Used where spectrum availability is limited.
   
- **FDD**: 
   - Best for legacy systems or applications with symmetric traffic (e.g., voice calls).
   - Suitable for environments with ample spectrum for separate UL/DL channels.

### **Conclusion**
TDD excels in flexibility and efficiency for data-driven, modern wireless networks, while FDD remains optimal for traditional, symmetric communication systems. The choice depends on the application, traffic patterns, and spectrum availability.
