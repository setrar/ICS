The three lines describe different wireless channel models:

1. **AWGN Model**: $ y = x + w $, where the channel has no fading and only adds white Gaussian noise.
2. **Indoor Fixed Channel**: $ y = h x + w $, with a constant channel gain $ h $, typical of stable indoor environments.
3. **Outdoor Fading Channel**: $ y = h x + w $, where $ h \sim \mathcal{CN}(0,1) $ (Rayleigh fading), rapidly changing $ 10^9 $ times per second, as is common in outdoor environments with high variability.

    <div style="margin-left: 40px;">
        <p>$\binom{y_1}{y_2} = \underline{h} . x + \underline{w} \implies \binom{y_1}{y_2} = \binom{h_1}{h_2} . x + \binom{w_1}{w_2} $ </p>
        <p>&emsp; $ \underline{y} = \underline{h} . x + \underline{w} $</p>
    </div>
</div>

In **Single Input Multiple Output (SIMO) systems**, let's break it down:

- $ \binom{y_1}{y_2} $ or $ \underline{y} $ is the received signal vector at two different receiver antennas.
- $ \underline{h} $ is the **channel vector**, representing the channel gains for each transmission path to the two receiver antennas. In this case, $ h_1 $ and $ h_2 $ are the channel coefficients for each receiver antenna.
- $ x $ is the **transmitted signal** from a single transmit antenna (hence "single input").
- $ \binom{w_1}{w_2} $ or $ \underline{w} $ is the noise vector at each receiver antenna, with $ w_1 $ and $ w_2 $ being noise components at receiver 1 and receiver 2, respectively.

In this case, you are dealing with a system where there is a single transmitter and multiple receivers (SIMO), where the same transmitted signal $ x $ is received at two antennas with different channel effects ($ h_1 $ and $ h_2 $), plus noise at each antenna.

In **MISO (Multiple Input Single Output) channel** , the expression:

$
y = \underline{h}^T \cdot \underline{x} + w
$

- $ y $ is the received signal.
- $ \underline{h}^T $ is the **channel vector**, representing the channel coefficients for each transmission path. In this case, the channel has two coefficients: $ h_1 $ and $ h_2 $.
- $ \underline{x} $ is the transmitted signal vector, where $ \binom{x_1}{x_2} $are the signals transmitted from two different antennas.
- $ w $ is the noise term, typically modeled as additive white Gaussian noise.

This is an expression for the received signal in a MISO system, where the signal received at a receiver antenna is a linear combination of the signals transmitted from two different transmit antennas, each affected by different channel coefficients $ h_1 $ and $ h_2 $, plus some noise.

<hr/>

<div style="display: flex; align-items: center;">
    <img src="images/CM_point_to_point4.png" alt="Point to Point 4" style="width: '10%'; height: '10%';">
    <div style="margin-left: 40px;">
        <p>$ \underline{y} = \binom{y_1}{y_2}, \quad \underline{x} = \binom{x_1}{x_2}, \quad H = \binom{\underline{h_1}^T}{\underline{h_2}^T}$ </p>
        <p>$ \binom{y_1}{y_2} = \binom{\underline{h_1}^T}{\underline{h_2}^T} \binom{x_1}{x_2} + \underline{w} \implies \underline{y} = H \underline{x} + \underline{w}$</p>
    </div>
</div>

This represents a MIMO (Multiple Input Multiple Output) channel, where both the input and output are vectors.

In the context of the equations:

1. **Definitions**:
   - $ \underline{y} = \binom{y_1}{y_2} $: This is the output vector containing multiple outputs.
   - $ \underline{x} = \binom{x_1}{x_2} $: This is the input vector containing multiple inputs.
   - $ H = \binom{\underline{h_1}^T}{\underline{h_2}^T} $: This is the channel matrix, where $ \underline{h_1}^T $ and $ \underline{h_2}^T $ are the transpose of the channel vectors corresponding to each output.

2. **Equation**:
   - The equation $ \underline{y} = H \underline{x} + \underline{w} $ indicates that the output vector $ \underline{y} $ is obtained by applying the channel matrix $ H $ to the input vector $ \underline{x} $, and then adding some noise $ \underline{w} $.

This setup is indicative of a MIMO system, where multiple inputs are transmitted to multiple outputs, allowing for more complex communication scenarios and improved data rates compared to MISO or SISO (Single Input Single Output) systems.

<div style="display: flex; align-items: center;">
    <img src="images/CM_point_to_point5.png" alt="Point to Point 5" style="width: '10%'; height: '10%';">
    <div style="margin-left: 40px;">
        <p>$ \underline{y} = ( H_1 \; H_2 ) \binom{x_1}{x_2} + \underline{w}$</p>
    </div>
</div>

This represents a MIMO (Multiple Input Multiple Output) system as well, but it has a specific structure that indicates a different arrangement of the channel matrix.

### Breakdown:

1. **Output Vector**:
   - $ \underline{y} $ is the output vector.

2. **Channel Matrix**:
   - $ ( H_1 \; H_2 ) $ suggests that the channel matrix is being constructed from two parts, $ H_1 $ and $ H_2 $. This implies that each output might be influenced by multiple inputs through different channels.

3. **Input Vector**:
   - $ \binom{x_1}{x_2} $: This is the input vector consisting of two components.

4. **Noise Term**:
   - $ \underline{w} $: This represents the noise added to the output.

### Interpretation:

- This formulation indicates that the outputs in $ \underline{y} $ are influenced by the inputs $ x_1 $ and $ x_2 $ through the channel matrices $ H_1 $ and $ H_2 $. It suggests a structure where each output may be a function of both inputs, possibly through linear combinations defined by the channel matrices.

Thus, overall, this formulation is also indicative of a MIMO system, where the outputs are derived from multiple inputs through a multi-channel configuration.
